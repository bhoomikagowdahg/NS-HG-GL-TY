```python
import csv
import pandas as pd

# Function to display inventory
def view_inventory():
    with open('item.csv', 'r') as file:
        print("Inventory:")
        print(file.read())

# Function to display inventory as CSV in Jupyter Notebook
def view_inventory_csv():
    inventory_data = []
    with open('item.csv', 'r') as file:
        inventory = csv.reader(file)
        for item in inventory:
            inventory_data.append(item)

    df = pd.DataFrame(inventory_data, columns=["Item Number", "Item Name", "Quantity", "Price"])
    display(df)

# Function to upload items to inventory
def upload_items():
    with open('item.csv', 'a') as file:
        item_number = input("Enter item number: ")
        item_name = input("Enter item name: ")
        quantity = input("Enter quantity: ")
        price = input("Enter price: ")
        file.write(f"{item_number},{item_name},{quantity},{price}\n")
    print("Item uploaded successfully.")

# Function to add items to cart
def add_to_cart():
    cart = []
    with open('item.csv', 'r') as file:
        inventory = csv.reader(file)
        for item in inventory:
            print(f"Item: {item[1]}, Price: {item[3]}")

    while True:
        item_name = input("Enter the name of the item you want to add to your cart (or 'done' to finish): ")
        if item_name.lower() == 'done':
            break
        quantity = input("Enter the quantity: ")
        cart.append((item_name, quantity))
    return cart

# Function to generate bill
def generate_bill(cart):
    total_price = 0
    print("Your Bill:")
    print("Item\t\tQuantity\tPrice")
    for item, quantity in cart:
        with open('item.csv', 'r') as file:
            inventory = csv.reader(file)
            for row in inventory:
                if row[1] == item:
                    price = float(row[3]) * int(quantity)
                    total_price += price
                    print(f"{item}{quantity}{price}")
    print(f"Total Price: {total_price}")

# Main function
def main():
    while True:
        print("\n1. User")
        print("2. Customer")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            while True:
                print("\n1. View Inventory")
                print("2. Upload Items")
                print("3. Exit")
                user_choice = input("Enter your choice: ")

                if user_choice == '1':
                    view_inventory()
                elif user_choice == '2':
                    upload_items()
                elif user_choice == '3':
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == '2':
            while True:
                print("\n1. View Inventory")
                print("2. Add to Cart")
                print("3. Generate Bill")
                print("4. Exit")
                customer_choice = input("Enter your choice: ")

                if customer_choice == '1':
                    view_inventory_csv()
                elif customer_choice == '2':
                    cart = add_to_cart()
                elif customer_choice == '3':
                    generate_bill(cart)
                elif customer_choice == '4':
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == '3':
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
